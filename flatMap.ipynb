```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setAppName('FlatMap')
sc = SparkContext.getOrCreate(conf=conf)
```

```python
rdd = sc.textFile('./sample.txt')
rdd.collect()
```

```python
mappedRDD = rdd.map(lambda x: x.split(' '))
mappedRDD.collect()
```

```python
mappedRDD1 = rdd.flatMap(lambda x: x.split(' '))
mappedRDD1.collect()
```
