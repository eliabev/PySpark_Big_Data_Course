```python
from pyspark.sql import SparkSession, functions as F

sc = SparkSession.builder\
                  .appName("Filter")\
                  .getOrCreate()
```

```python
rdd = sc.read.text('/Volumes/workspace/default/volume/sample.txt')

filtered_column = rdd.filter(rdd.value != '12 12 33')

filtered_column.collect()
```

```python
[row.value for row in filtered_column.collect()]
```

```python
quiz_txt = sc.read.text('/Volumes/workspace/default/volume/quiz2.txt')

filtered_result = quiz_txt.withColumn('test_values', F.explode(F.split(F.col('value'), ' '))).filter(~F.substring(F.col('test_values'), 1, 1).isin('a', 'c')).select('test_values')

filtered_result.collect()
```
