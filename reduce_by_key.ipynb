```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setAppName('Reduce By Key')
sc = SparkContext.getOrCreate(conf)

rdd = sc.textFile('./sample.txt')
```

```python
rdd.collect()
```

```python
rdd2 = rdd.flatMap(lambda x: x.split(' '))\
    .map(lambda x: (x, 1))

rdd2.collect()
```

```python
def reduce_func(acc, x):
    #print(acc, x, '\n')
    return acc + x

rdd2.reduceByKey(reduce_func).collect()
```

```python
quiz_txt = sc.textFile('quiz3.txt')
quiz_txt.collect()
```

```python
quiz_txt.flatMap(lambda x: x.split(' '))\
        .map(lambda x: (x, 1))\
        .reduceByKey(lambda acc, x: acc + x)\
        .sortBy(lambda x: (-x[1], x[0]), ascending=True)\
        .collect()
```
