```python
from pyspark import SparkContext, SparkConf

conf = SparkConf().setAppName('Filter Quiz')
sc = SparkContext.getOrCreate(conf=conf)
```

```python
quiz_text = sc.textFile('quiz2.txt')

filtered_result = quiz_text.flatMap(lambda x: x.split(' ')).filter(lambda x: x[0] not in ('a', 'c'))
filtered_result.collect()
```

```python
def split_filter(x):
    y = list(filter(lambda x: x[0] not in ('a', 'c'), x.split(' ')))
    return y if len(y) > 0 else None


filtered_result1 = quiz_text.map(split_filter).filter(lambda x: x != None)
filtered_result1.collect()
```
