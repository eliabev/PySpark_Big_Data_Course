```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setAppName("Read File")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
text = sc.textFile('./sample.txt')
```

```python
text.collect()
```

```python
def strip_split_sum(x):
    stripped_x = x.strip().split()
    return sum(map(int, stripped_x))
```

```python
rdd_sum = text.map(strip_split_sum)
              
```

```python
rdd_sum.collect()
```

## Quiz

```python
def count_word_in_lines(x):
    splitted = x.split(' ')
    word_count = []
    for s in splitted:
        word_count.append(len(s))
    
    return word_count
```

```python
quiz_txt = sc.textFile('quiz.txt')
print(quiz_txt.collect())
quiz_txt.map(count_word_in_lines).collect()
```

```python
quiz_txt.map(lambda x: [len(w) for w in x.split(' ')]).collect()
```
