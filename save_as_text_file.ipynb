```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setAppName('Save as Text File')
sc = SparkContext.getOrCreate(conf)

rdd = sc.textFile('quiz3.txt')
```

```python
"""
It won't create a text file, but a folder.
This folder will contain the result of all the partitions (2, in this case)
but not a single file result
"""
rdd.saveAsTextFile('./example.txt')
```
