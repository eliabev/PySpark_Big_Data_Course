```python
from pyspark.sql import SparkSession, functions as F

sc = SparkSession.builder\
                  .appName("Group By Key")\
                  .getOrCreate()
```

```python
text_df = sc.read.text('/Volumes/workspace/default/volume/quiz2.txt')

mapped = text_df.select(F.struct(F.col('value'), F.lit(F.length(F.col('value')))))

mapped.collect()
```

```python
tuple_df = text_df.select(
        F.col('value'),
        F.size(F.split(F.col('value'), ' ')).alias('word_count')
    )
tuple_df.collect()
```

```python
flatted_df = text_df.select(
        F.explode(F.split(F.col('value'), ' ')).alias('word'),
        F.length(F.col('word')).alias('count')
    )
flatted_df.collect()
```

```python
flatted_df = flatted_df.groupBy('word').agg(F.collect_list('count').alias('grouped'))
```

```python
[(Row.word, Row.grouped) for Row in flatted_df.collect()]
```
