```python
from pyspark.sql import SparkSession, functions as F

sc = SparkSession.builder\
                  .appName("Distinct")\
                  .getOrCreate()
```

```python
text_df = sc.read.text('/Volumes/workspace/default/volume/sample.txt')

distinct_df = text_df.select(F.explode(F.split(text_df['value'], ' ')).alias('number')).distinct()
distinct_df.collect()
```
