```python
from pyspark.sql import SparkSession, functions as F

sc = SparkSession.builder\
                 .appName("FlatMap")\
                 .getOrCreate()
```

```python
rdd = sc.read.text('/Volumes/workspace/default/volume/sample.txt')

flatMappedRdd = rdd.withColumn('flatMapped', F.explode(F.split(F.trim('value'), ' ')))
flatMappedRdd.select('flatMapped').show()
```
