```python
from pyspark.sql import functions as F, SparkSession

spark = SparkSession.builder.appName("Reduce by Key").getOrCreate()

text_df = spark.read.text('/Volumes/workspace/default/volume/sample.txt')
text_df.collect()
```

```python
word_count_df = text_df\
    .select(F.explode(F.split(F.col('value'), ' ')).alias('word'))\
    .groupBy('word').agg(F.count('*').alias('count'))

display(word_count_df)
```
